**1. Duration_ms ve Views Arasındaki Korelasyon:**

Korelasyon değeri: 0.021693  
Bu çok düşük bir korelasyon değeri olduğu için, şarkı süresi ile izlenme sayısı arasında çok zayıf bir ilişki vardır. Yani, şarkının uzunluğu ile izlenme sayısı arasında anlamlı bir ilişki gözlemlenmiyor.  
Pozitif korelasyon olduğundan, çok küçük bir dereceyle de olsa, uzun şarkılar ile daha fazla izlenme sayısı olma ihtimali olabilir, ancak bu ilişki pratikte çok zayıf.

**2. Duration_ms ve Likes Arasındaki Korelasyon:**

Korelasyon değeri: 0.005555  
Bu da çok düşük bir pozitif korelasyon değeri olup, şarkı süresi ile beğeni sayısı arasında çok zayıf bir ilişki olduğunu gösterir. Yani, şarkı süresi ile beğeni sayısı arasında neredeyse hiçbir ilişki bulunmamaktadır.

**3. Views ve Likes Arasındaki Korelasyon:**

Korelasyon değeri: 0.891101  
Bu, izlenme sayısı ile beğeni sayısı arasında güçlü bir pozitif ilişki olduğunu gösterir. Yani, bir şarkı ne kadar fazla izlenirse, büyük olasılıkla o kadar fazla beğeni alır. Bu iki değişken arasında güçlü bir doğrusal ilişki vardır.

**Sonuçlar:**

- Şarkı süresi ile popülerlik (izlenme ve beğeni sayıları) arasında belirgin bir ilişki bulunmamaktadır.  
- İzlenme sayıları ile beğeni sayıları arasındaki ilişki çok güçlüdür; şarkılar daha fazla izlendikçe, daha fazla beğeni alıyor gibi görünüyor.


```python
# Belirtilen sütunlardaki eksik (null) değerleri içeren satırları kalıcı olarak silmek
df.dropna(subset=['Danceability', 'Energy', 'Key', 'Loudness', 'Speechiness', 'Acousticness', 
                  'Instrumentalness', 'Liveness', 'Valence', 'Tempo', 'Duration_ms'], inplace=True)
```

```python
df.info()
df.describe()
```

```python
#işime yaramayan sütunları siliyorum.
df.drop(['Unnamed: 0'], axis=1, inplace=True)
df.drop(['Url_spotify'], axis=1, inplace=True)
df.drop(['Uri'], axis=1, inplace=True)
df.drop(['Url_youtube'], axis=1, inplace=True)
df.drop(['Title'], axis=1, inplace=True)
df.drop(['Description'], axis=1, inplace=True)
```

```python
#veri bütünlüğü korumak için ve analizlerin daha doğru sonuç vermesi için null değerleri dropluyorum.
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
# Duration_ms'yi dakikaya çevirmek için yeni bir sütun ekleyelim
df['Duration_min'] = df['Duration_ms'] / 1000 / 60
print(df[['Track', 'Duration_ms', 'Duration_min']].head())
```

```python
# Danceability ve Energy sütunları için histogramlar
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
sns.histplot(df['Danceability'], kde=True, color='blue', bins=30)
plt.title('Danceability Dağılımı')

plt.subplot(1, 2, 2)
sns.histplot(df['Energy'], kde=True, color='red', bins=30)
plt.title('Energy Dağılımı')

plt.tight_layout()
plt.show()
```

**Aykırı değerleri inceleme**

```python
from scipy import stats


# Sayısal sütunları seçiyoruz
numeric_columns = df.select_dtypes(include=[np.number]).columns

# Z-skoru hesaplama
z_scores = np.abs(stats.zscore(df[numeric_columns]))

# Z-skoru > 3 olan aykırı değerler
outliers_zscore = (z_scores > 3).any(axis=1)

# Aykırı değerleri gösterme
outliers_zscore_data = df[outliers_zscore]

# Aykırı değerlerin sayısı
print(f"Aykırı değer sayısı (Z-Score): {outliers_zscore_data.shape[0]}")
print(outliers_zscore_data)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

for column in numeric_columns:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=df[column])
    plt.title(f"Boxplot of {column}")
    plt.show()
```

# SANATÇI-ŞARKI-MÜZİK TÜRÜ VE DİĞER ETKİLEŞİMLER

**En Çok Dinlenen İlk 10 Sanatçı**

```python
# Sanatçılar bazında toplam dinlenme sayısını hesaplayalım
top_artists = df.groupby('Artist')['Views'].sum().reset_index()

# Dinlenme sayısına göre azalan sıralama yapalım
top_artists_sorted = top_artists.sort_values(by='Views', ascending=False)

# En çok dinlenen ilk 10 sanatçıyı alalım
top_10_artists = top_artists_sorted.head(10)

# Sonuçları yazdıralım
print("En çok dinlenen ilk 10 sanatçı ve toplam dinlenme sayıları:\n")
print(top_10_artists)

```

```python
# Görselleştirme için matplotlib ve seaborn ayarlarını yapalım
plt.figure(figsize=(12, 8))

# Barplot ile en çok dinlenen ilk 10 sanatçıyı görselleştirelim
sns.barplot(x='Views', y='Artist', data=top_10_artists, palette='viridis')

# Çubukların üzerine dinlenme sayılarını ekleyelim
for i in range(top_10_artists.shape[0]):
    plt.text(
        top_10_artists['Views'].iloc[i] + 0.05,  # X konumu (çubuğun biraz sağında)
        i,  # Y konumu (yani sanatçının bulunduğu satır)
        f"{top_10_artists['Views'].iloc[i]:,.0f}",  # Dinlenme sayısını formatlayarak yazalım
        ha='left',  # Etiket hizası
        va='center',  # Dikey hizalama
        fontsize=10
    )

# Başlık ve etiketler ekleyelim
plt.title('En Çok Dinlenen İlk 10 Sanatçı(izlenme Sayısına Göre)', fontsize=16)
plt.xlabel('Toplam Dinlenme Sayısı (Views)', fontsize=12)
plt.ylabel('Sanatçılar', fontsize=12)

# Görselleştirmeyi gösterelim
plt.tight_layout()
plt.show()
```

**En Çok Dinlenen İlk 10 Sanatçı-Şarkı-Albüm Türü-İzlenme Sayısı**

```python
# Sanatçı, şarkı, albüm türü ve dinlenme sayısını içeren sütunları kullanarak en çok dinlenen sanatçıları bulalım
top_artists = df.groupby(['Artist', 'Track', 'Album_type'])['Views'].sum().reset_index()

# Dinlenme sayısına göre azalan sıralama yapalım
top_artists_sorted = top_artists.sort_values(by='Views', ascending=False)

# En çok dinlenen ilk 10 sanatçı ve şarkısını alalım
top_10_artists = top_artists_sorted.head(10)

# Sonuçları yazdıralım
print("En çok dinlenen ilk 10 sanatçı ve şarkıları:\n")
print(top_10_artists[['Artist', 'Track', 'Album_type', 'Views']])
```

```python
plt.figure(figsize=(12, 8))

sns.barplot(x='Views', y='Artist', data=top_10_artists, palette='viridis')

for i in range(top_10_artists.shape[0]):
    plt.text(
        top_10_artists['Views'].iloc[i] + 0.05,  # X konumu (çubuğun biraz sağında)
        i,  # Y konumu (yani sanatçının bulunduğu satır)
        f"{top_10_artists['Track'].iloc[i]} ({top_10_artists['Album_type'].iloc[i]})",  # Şarkı adı ve albüm türü
        ha='left',  # Etiket hizası
        va='center',  # Dikey hizalama
        fontsize=9
    )

plt.title('En Çok Dinlenen İlk 10 Sanatçı ve Şarkıları', fontsize=16)
plt.xlabel('Toplam Dinlenme Sayısı (Views)', fontsize=12)
plt.ylabel('Sanatçılar', fontsize=12)

plt.tight_layout()
plt.show()
```
