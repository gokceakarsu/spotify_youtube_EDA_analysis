**AykÄ±rÄ± deÄŸerleri inceleme**

```python
from scipy import stats


# SayÄ±sal sÃ¼tunlarÄ± seÃ§iyoruz
numeric_columns = df.select_dtypes(include=[np.number]).columns

# Z-skoru hesaplama
z_scores = np.abs(stats.zscore(df[numeric_columns]))

# Z-skoru > 3 olan aykÄ±rÄ± deÄŸerler
outliers_zscore = (z_scores > 3).any(axis=1)

# AykÄ±rÄ± deÄŸerleri gÃ¶sterme
outliers_zscore_data = df[outliers_zscore]

# AykÄ±rÄ± deÄŸerlerin sayÄ±sÄ±
print(f"AykÄ±rÄ± deÄŸer sayÄ±sÄ± (Z-Score): {outliers_zscore_data.shape[0]}")
print(outliers_zscore_data)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

for column in numeric_columns:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=df[column])
    plt.title(f"Boxplot of {column}")
    plt.show()
```

# SANATÃ‡I-ÅžARKI-MÃœZÄ°K TÃœRÃœ VE DÄ°ÄžER ETKÄ°LEÅžÄ°MLER

**En Ã‡ok Dinlenen Ä°lk 10 SanatÃ§Ä±**

```python
# SanatÃ§Ä±lar bazÄ±nda toplam dinlenme sayÄ±sÄ±nÄ± hesaplayalÄ±m
top_artists = df.groupby('Artist')['Views'].sum().reset_index()

# Dinlenme sayÄ±sÄ±na gÃ¶re azalan sÄ±ralama yapalÄ±m
top_artists_sorted = top_artists.sort_values(by='Views', ascending=False)

# En Ã§ok dinlenen ilk 10 sanatÃ§Ä±yÄ± alalÄ±m
top_10_artists = top_artists_sorted.head(10)

# SonuÃ§larÄ± yazdÄ±ralÄ±m
print("En Ã§ok dinlenen ilk 10 sanatÃ§Ä± ve toplam dinlenme sayÄ±larÄ±:\n")
print(top_10_artists)

```

```python
# GÃ¶rselleÅŸtirme iÃ§in matplotlib ve seaborn ayarlarÄ±nÄ± yapalÄ±m
plt.figure(figsize=(12, 8))

# Barplot ile en Ã§ok dinlenen ilk 10 sanatÃ§Ä±yÄ± gÃ¶rselleÅŸtirelim
sns.barplot(x='Views', y='Artist', data=top_10_artists, palette='viridis')

# Ã‡ubuklarÄ±n Ã¼zerine dinlenme sayÄ±larÄ±nÄ± ekleyelim
for i in range(top_10_artists.shape[0]):
    plt.text(
        top_10_artists['Views'].iloc[i] + 0.05,  # X konumu (Ã§ubuÄŸun biraz saÄŸÄ±nda)
        i,  # Y konumu (yani sanatÃ§Ä±nÄ±n bulunduÄŸu satÄ±r)
        f"{top_10_artists['Views'].iloc[i]:,.0f}",  # Dinlenme sayÄ±sÄ±nÄ± formatlayarak yazalÄ±m
        ha='left',  # Etiket hizasÄ±
        va='center',  # Dikey hizalama
        fontsize=10
    )

# BaÅŸlÄ±k ve etiketler ekleyelim
plt.title('En Ã‡ok Dinlenen Ä°lk 10 SanatÃ§Ä±(izlenme SayÄ±sÄ±na GÃ¶re)', fontsize=16)
plt.xlabel('Toplam Dinlenme SayÄ±sÄ± (Views)', fontsize=12)
plt.ylabel('SanatÃ§Ä±lar', fontsize=12)

# GÃ¶rselleÅŸtirmeyi gÃ¶sterelim
plt.tight_layout()
plt.show()
```

**En Ã‡ok Dinlenen Ä°lk 10 SanatÃ§Ä±-ÅžarkÄ±-AlbÃ¼m TÃ¼rÃ¼-Ä°zlenme SayÄ±sÄ±**

```python
# SanatÃ§Ä±, ÅŸarkÄ±, albÃ¼m tÃ¼rÃ¼ ve dinlenme sayÄ±sÄ±nÄ± iÃ§eren sÃ¼tunlarÄ± kullanarak en Ã§ok dinlenen sanatÃ§Ä±larÄ± bulalÄ±m
top_artists = df.groupby(['Artist', 'Track', 'Album_type'])['Views'].sum().reset_index()

# Dinlenme sayÄ±sÄ±na gÃ¶re azalan sÄ±ralama yapalÄ±m
top_artists_sorted = top_artists.sort_values(by='Views', ascending=False)

# En Ã§ok dinlenen ilk 10 sanatÃ§Ä± ve ÅŸarkÄ±sÄ±nÄ± alalÄ±m
top_10_artists = top_artists_sorted.head(10)

# SonuÃ§larÄ± yazdÄ±ralÄ±m
print("En Ã§ok dinlenen ilk 10 sanatÃ§Ä± ve ÅŸarkÄ±larÄ±:\n")
print(top_10_artists[['Artist', 'Track', 'Album_type', 'Views']])
```

```python
plt.figure(figsize=(12, 8))

sns.barplot(x='Views', y='Artist', data=top_10_artists, palette='viridis')

for i in range(top_10_artists.shape[0]):
    plt.text(
        top_10_artists['Views'].iloc[i] + 0.05,  # X konumu (Ã§ubuÄŸun biraz saÄŸÄ±nda)
        i,  # Y konumu (yani sanatÃ§Ä±nÄ±n bulunduÄŸu satÄ±r)
        f"{top_10_artists['Track'].iloc[i]} ({top_10_artists['Album_type'].iloc[i]})",  # ÅžarkÄ± adÄ± ve albÃ¼m tÃ¼rÃ¼
        ha='left',  # Etiket hizasÄ±
        va='center',  # Dikey hizalama
        fontsize=9
    )

plt.title('En Ã‡ok Dinlenen Ä°lk 10 SanatÃ§Ä± ve ÅžarkÄ±larÄ±', fontsize=16)
plt.xlabel('Toplam Dinlenme SayÄ±sÄ± (Views)', fontsize=12)
plt.ylabel('SanatÃ§Ä±lar', fontsize=12)

plt.tight_layout()
plt.show()
```
